# Power spectral density of data

This notebook is an example of taking a power spectral density
of data. The power spectral density of a signal containing
.5 Hz and 1.59 Hz signals is plotted.

```python
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
sp.set_printoptions(precision=4, linewidth=250, suppress=True)
```

```python
t = sp.arange(0, 100.1, 0.1)
```

```python
y = sp.sin(sp.pi*t) + sp.sin(10*t)
```

```python
x = sp.fft(y, 512)
```

$S_{xx}(\omega) = \lim_{t \to \infty} {\bf E} \big[|x_T (\omega)|^2\big]$

```python
Pyy = sp.real(x*sp.conj(x)/512)
```

```python
f = 10*sp.arange(256)/512
```

```python
fig = plt.figure(figsize=(8,6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)
fig.tight_layout(h_pad=4)

ax1.set_title('sin(pi*t)+sin(10*t) versus Time')
ax1.set_xlabel('Time')
ax1.set_ylabel('F(t)')

ax2.set_title('Power Spectral Density of F(t)')
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Power Spectral Density')

ax1.plot(t, y)
ax2.plot(f, Pyy[:256])
```
